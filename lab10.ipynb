```python
import requests
from bs4 import BeautifulSoup
import csv

def fetch_bbc_news():
    url = 'https://marathi.abplive.com/'
    response = requests.get(url)

    # Перевірте статус відповіді; 200 означає, що все гаразд
    if response.status_code != 200:
        print(f'Failed to retrieve the page with status code: {response.status_code}')
        return None

    soup = BeautifulSoup(response.content, 'html.parser')

    # Збір заголовків новин
    headlines = []
    for article in soup.find_all('p'):
        headline_text = article.text.strip()
        if headline_text:  # Перевірка на наявність тексту
            headlines.append(headline_text)

    # Збереження заголовків новин у файл CSV
    with open('marathi_headlines.csv', mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(['Headline'])  # Заголовок стовпця
        for headline in headlines:
            writer.writerow([headline])

# Виклик функції
fetch_bbc_news()
```

```python
import csv

# Ім'я CSV-файлу
csv_filename = 'marathi_headlines.csv'

# Змінна для збереження вмісту файлу
marathi_text = ''

# Відкриваємо CSV-файл для читання
with open(csv_filename, mode='r', encoding='utf-8') as file:
    reader = csv.reader(file)

    # Ітеруємося через рядки CSV-файлу та додаємо їх до змінної text
    for row in reader:
        marathi_text += ', '.join(row) + '\n'

# Виводимо вміст
print(marathi_text)
```

```python
from googletrans import Translator

def translate_to_english(marath_text):
    translator = Translator()
    translated_text_english = translator.translate(marathi_text, src='mr', dest='en')
    return translated_text_english.text



# Переклад тексту на українську
english_text = translate_to_english(marathi_text)

print("Маратхі текст:", marathi_text)
print("Переклад на англійську:", english_text)
# Видаляємо слово "заголовки" з тексту
english_text = english_text.replace("заголовки", "")
```

```python
from googletrans import Translator

def translate_to_ukrainian(text):
    translator = Translator()
    try:
        translated_text = translator.translate(text, src='en', dest='uk')
        return translated_text.text
    except Exception as e:
        print(f"An error occurred during translation: {e}")
        return None



# Виклик функції для перекладу англійського тексту на українську
text = translate_to_ukrainian(english_text)

# Перевірка, чи переклад виконано успішно і виведення результату
if text:
    print("Англійський текст:", english_text, '\n', '\n')
    print("Переклад на українську:", text)
else:
    print("Не вдалось перекласти текст.")
# Видаляємо слово "заголовки" з тексту
text = text.replace("заголовки", "")
text = text.replace("Заголовок", "")
text = text.replace("з", "")
text = text.replace("в", "")
text = text.replace("до", "")
```

```python

# length of text ( includes spaces)
print("length of text: ",len(text))

# split the text
splitted_text = text.split() # default split methods splits text according to spaces
print("Splitted text: ",splitted_text)    # splitted_text is a list that includes words of text sentence
# each word is called token in text maning world.
```

```python
# find specific words with list comprehension method
specific_words = [word for word in splitted_text if(len(word)>2)]
print("Words which are more than 3 letter: ",specific_words)

# capitalized words with istitle() method that finds capitalized words
capital_words = [ word for word in splitted_text if word.istitle()]
print("Capitalized words: ",capital_words)

# words which end with "o": endswith() method finds last letter of word
words_end_with_o =  [word for word in splitted_text if word.endswith("o")]
print("words end with o: ",words_end_with_o)

# words which starts with "w": startswith() method
words_start_with_w = [word for word in splitted_text if word.startswith("w")]
print("words start with w: ",words_start_with_w)
```

```python
# unique with set() method
print("unique words: ",set(splitted_text))  # actually the word "no" is occured twice bc one word is "no" and others "No" there is a capital letter at first letter

# make all letters lowercase with lower() method
lowercase_text = [word.lower() for word in splitted_text]

# then find uniques again with set() method
print("unique words: ",set(lowercase_text))
```

```python
import re

def clean_text(text):
    # Вилучення всіх символів, крім літер українського алфавіту, літер "і", "ї" та пробілів
    cleaned_text = re.sub(r'[^А-Яа-яІіЇї\s]', '', text)
    return cleaned_text


cleaned_text = clean_text(text)

print(cleaned_text)

```

```python
from nltk.stem import WordNetLemmatizer

import nltk
nltk.download('wordnet')


def lemmatize_text(cleaned_text):
    lemmatizer = WordNetLemmatizer()
    words = word_tokenize(cleaned_text)
    lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
    lemmatized_text = ' '.join(lemmatized_words)
    return lemmatized_text

lemmatized_text = lemmatize_text(cleaned_text)
print(lemmatized_text)
```

```python

# Розділення тексту на слова
original_words = text.split()
processed_words = lemmatized_text.split()

# Кількість слів в оригінальному та обробленому тексті
original_word_count = len(original_words)
processed_word_count = len(processed_words)
```

```python
import spacy
import matplotlib.pyplot as plt

# Переконайтеся, що ви вже встановили модель 'uk_core_news_trf', як було роз'яснено раніше
nlp = spacy.load('uk_core_news_trf')

# Обробка тексту spaCy

doc = nlp(text)

# Підрахунок відносних частот слів
word_freq = {}
total_words = len(doc)

for token in doc:
    word = token.text.lower()  # Перетворення слова до нижнього регістру
    if word.isalpha():  # Ігноруємо пунктуацію
        if word in word_freq:
            word_freq[word] += 1
        else:
            word_freq[word] = 1

# Сортування словник у порядку спадання відносних частот
sorted_word_freq = dict(sorted(word_freq.items(), key=lambda item: item[1], reverse=True))

# Відібрані слова і їх відносні частоти
selected_words = list(sorted_word_freq.keys())[:10]  # Відібрати топ-10 слів
selected_freqs = [sorted_word_freq[word] / total_words for word in selected_words]

# Побудова графіка відносних частот слів
plt.figure(figsize=(10, 6))
plt.bar(selected_words, selected_freqs, color='green')
plt.xlabel('Слова')
plt.ylabel('Відносні частоти')
plt.title('Графік відносних частот слів')
plt.xticks(rotation=45)  # Поворот підписів на 45 градусів, щоб зробити їх більш читабельними
plt.show()
```

```python
import spacy
import networkx as nx
import matplotlib.pyplot as plt

# Завантаження моделі spaCy для української мови
nlp = spacy.load('uk_core_news_trf')

# Обробка тексту spaCy
doc = nlp(text)

# Створення графа
G = nx.Graph()

# Додавання слів як вузлів до графа
for token in doc:
    if token.is_alpha:  # Ігноруємо пунктуацію
        G.add_node(token.text)

# Додавання зв'язків між словами на основі синтаксичних зв'язків
for token in doc:
    for child in token.children:
        if token.is_alpha and child.is_alpha:
            G.add_edge(token.text, child.text)

# Побудова графа
pos = nx.spring_layout(G, seed=42)  # Встановлюємо seed для стабільності розташування вузлів
plt.figure(figsize=(18, 10))
nx.draw(G, pos, with_labels=True, node_size=1000, node_color='skyblue', font_size=8, font_color='black')
plt.title('Мережа слів')
plt.show()
```

```python
# Отримання списку вузлів та зв'язків
nodes = list(G.nodes)
edges = list(G.edges)

# Виведення вузлів
print("Список вузлів:")
for node in nodes:
    print(node)

# Виведення зв'язків
print("\nСписок зв'язків:")
for edge in edges:
    print(edge)
```
